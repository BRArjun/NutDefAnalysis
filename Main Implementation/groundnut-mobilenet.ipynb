```python
# Complete Groundnut Plant Leaf Dataset Analysis
# Comprehensive analysis and modeling pipeline for plant leaf classification

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

class GroundnutLeafAnalyzer:
    def __init__(self, base_path="/kaggle/input/groundnut-plant-leaf-data/Dataset of groundnut plant leaf images for classification and detection/Groundnut_Leaf_dataset"):
        self.base_path = base_path
        self.train_path = os.path.join(base_path, "train")
        self.test_path = os.path.join(base_path, "test")
        self.img_size = (224, 224)
        self.batch_size = 32
        self.classes = []
        self.class_counts = {}
        self.model = None
        self.history = None
        
    def explore_dataset_structure(self):
        """Explore and visualize dataset structure"""
        print("=== DATASET STRUCTURE EXPLORATION ===")
        print(f"Base path: {self.base_path}")
        print(f"Train path: {self.train_path}")
        print(f"Test path: {self.test_path}")
        
        # Check if paths exist
        if not os.path.exists(self.base_path):
            print(f"Error: Base path {self.base_path} does not exist!")
            return
            
        # Get class directories
        if os.path.exists(self.train_path):
            self.classes = [d for d in os.listdir(self.train_path) 
                           if os.path.isdir(os.path.join(self.train_path, d))]
            print(f"\nFound {len(self.classes)} classes: {self.classes}")
        
        # Count samples per class
        train_counts = {}
        test_counts = {}
        
        for class_name in self.classes:
            train_class_path = os.path.join(self.train_path, class_name)
            test_class_path = os.path.join(self.test_path, class_name)
            
            if os.path.exists(train_class_path):
                train_counts[class_name] = len([f for f in os.listdir(train_class_path) 
                                              if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
            
            if os.path.exists(test_class_path):
                test_counts[class_name] = len([f for f in os.listdir(test_class_path) 
                                             if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
        
        self.class_counts = {'train': train_counts, 'test': test_counts}
        
        # Display counts
        print("\n=== CLASS DISTRIBUTION ===")
        df_counts = pd.DataFrame({
            'Class': self.classes,
            'Train_Count': [train_counts.get(c, 0) for c in self.classes],
            'Test_Count': [test_counts.get(c, 0) for c in self.classes]
        })
        df_counts['Total'] = df_counts['Train_Count'] + df_counts['Test_Count']
        print(df_counts)
        
        return df_counts
    
    def visualize_class_distribution(self, df_counts):
        """Create visualizations for class distribution"""
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        
        # Train distribution
        axes[0, 0].bar(df_counts['Class'], df_counts['Train_Count'], color='skyblue')
        axes[0, 0].set_title('Training Set Distribution')
        axes[0, 0].set_xlabel('Classes')
        axes[0, 0].set_ylabel('Number of Images')
        axes[0, 0].tick_params(axis='x', rotation=45)
        
        # Test distribution
        axes[0, 1].bar(df_counts['Class'], df_counts['Test_Count'], color='lightcoral')
        axes[0, 1].set_title('Test Set Distribution')
        axes[0, 1].set_xlabel('Classes')
        axes[0, 1].set_ylabel('Number of Images')
        axes[0, 1].tick_params(axis='x', rotation=45)
        
        # Total distribution pie chart
        axes[1, 0].pie(df_counts['Total'], labels=df_counts['Class'], autopct='%1.1f%%')
        axes[1, 0].set_title('Overall Class Distribution')
        
        # Train vs Test comparison
        x = np.arange(len(df_counts['Class']))
        width = 0.35
        axes[1, 1].bar(x - width/2, df_counts['Train_Count'], width, label='Train', color='skyblue')
        axes[1, 1].bar(x + width/2, df_counts['Test_Count'], width, label='Test', color='lightcoral')
        axes[1, 1].set_xlabel('Classes')
        axes[1, 1].set_ylabel('Number of Images')
        axes[1, 1].set_title('Train vs Test Distribution')
        axes[1, 1].set_xticks(x)
        axes[1, 1].set_xticklabels(df_counts['Class'], rotation=45)
        axes[1, 1].legend()
        
        plt.tight_layout()
        plt.show()
        
        # Print summary statistics
        print("\n=== DATASET SUMMARY STATISTICS ===")
        print(f"Total images: {df_counts['Total'].sum()}")
        print(f"Training images: {df_counts['Train_Count'].sum()}")
        print(f"Test images: {df_counts['Test_Count'].sum()}")
        print(f"Number of classes: {len(self.classes)}")
        print(f"Average images per class: {df_counts['Total'].mean():.1f}")
        print(f"Standard deviation: {df_counts['Total'].std():.1f}")
    
    def sample_images_visualization(self, samples_per_class=3):
        """Display sample images from each class"""
        if not self.classes:
            print("No classes found. Run explore_dataset_structure() first.")
            return
            
        fig, axes = plt.subplots(len(self.classes), samples_per_class, 
                                figsize=(samples_per_class*3, len(self.classes)*3))
        
        if len(self.classes) == 1:
            axes = axes.reshape(1, -1)
        
        for i, class_name in enumerate(self.classes):
            class_path = os.path.join(self.train_path, class_name)
            if os.path.exists(class_path):
                image_files = [f for f in os.listdir(class_path) 
                              if f.lower().endswith(('.png', '.jpg', '.jpeg'))][:samples_per_class]
                
                for j, img_file in enumerate(image_files):
                    img_path = os.path.join(class_path, img_file)
                    img = load_img(img_path, target_size=self.img_size)
                    
                    if len(self.classes) > 1:
                        axes[i, j].imshow(img)
                        axes[i, j].set_title(f"{class_name}\n{img_file}")
                        axes[i, j].axis('off')
                    else:
                        axes[j].imshow(img)
                        axes[j].set_title(f"{class_name}\n{img_file}")
                        axes[j].axis('off')
        
        plt.tight_layout()
        plt.show()
    
    def analyze_image_properties(self):
        """Analyze image properties like dimensions, file sizes, etc."""
        print("\n=== IMAGE PROPERTIES ANALYSIS ===")
        
        dimensions = []
        file_sizes = []
        formats = []
        
        for class_name in self.classes:
            class_path = os.path.join(self.train_path, class_name)
            if os.path.exists(class_path):
                image_files = [f for f in os.listdir(class_path) 
                              if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
                
                for img_file in image_files[:50]:  # Sample first 50 images per class
                    img_path = os.path.join(class_path, img_file)
                    try:
                        with Image.open(img_path) as img:
                            dimensions.append(img.size)
                            formats.append(img.format)
                        file_sizes.append(os.path.getsize(img_path))
                    except Exception as e:
                        print(f"Error processing {img_path}: {e}")
        
        # Convert to DataFrame for analysis
        df_props = pd.DataFrame({
            'Width': [d[0] for d in dimensions],
            'Height': [d[1] for d in dimensions],
            'FileSize_KB': [s/1024 for s in file_sizes],
            'Format': formats
        })
        
        print("\nImage Dimensions Statistics:")
        print(df_props[['Width', 'Height', 'FileSize_KB']].describe())
        
        print(f"\nImage Formats: {df_props['Format'].value_counts().to_dict()}")
        
        # Visualize dimension distribution
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))
        
        axes[0].hist(df_props['Width'], bins=30, alpha=0.7)
        axes[0].set_title('Width Distribution')
        axes[0].set_xlabel('Width (pixels)')
        
        axes[1].hist(df_props['Height'], bins=30, alpha=0.7)
        axes[1].set_title('Height Distribution')
        axes[1].set_xlabel('Height (pixels)')
        
        axes[2].hist(df_props['FileSize_KB'], bins=30, alpha=0.7)
        axes[2].set_title('File Size Distribution')
        axes[2].set_xlabel('File Size (KB)')
        
        plt.tight_layout()
        plt.show()
        
        return df_props
    
    def create_data_generators(self, validation_split=0.2):
        """Create data generators with augmentation"""
        print("\n=== CREATING DATA GENERATORS ===")
        
        # Training data generator with augmentation
        train_datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            fill_mode='nearest',
            validation_split=validation_split
        )
        
        # Test data generator (only rescaling)
        test_datagen = ImageDataGenerator(rescale=1./255)
        
        # Create generators
        train_generator = train_datagen.flow_from_directory(
            self.train_path,
            target_size=self.img_size,
            batch_size=self.batch_size,
            class_mode='categorical',
            subset='training'
        )
        
        validation_generator = train_datagen.flow_from_directory(
            self.train_path,
            target_size=self.img_size,
            batch_size=self.batch_size,
            class_mode='categorical',
            subset='validation'
        )
        
        test_generator = test_datagen.flow_from_directory(
            self.test_path,
            target_size=self.img_size,
            batch_size=self.batch_size,
            class_mode='categorical',
            shuffle=False
        )
        
        print(f"Training samples: {train_generator.samples}")
        print(f"Validation samples: {validation_generator.samples}")
        print(f"Test samples: {test_generator.samples}")
        print(f"Number of classes: {train_generator.num_classes}")
        print(f"Class indices: {train_generator.class_indices}")
        
        return train_generator, validation_generator, test_generator
    
    def build_mobilenet_model(self, num_classes, trainable_layers=0):
        """Build MobileNet-based model with transfer learning
        
        Args:
            num_classes: Number of output classes
            trainable_layers: Number of top layers to make trainable (0 = freeze all base layers)
        """
        # Load MobileNet base model
        base_model = MobileNet(
            weights='imagenet',
            include_top=False,
            input_shape=(*self.img_size, 3)
        )
        
        # Freeze base model layers
        base_model.trainable = False
        
        # Optionally make some top layers trainable for fine-tuning
        if trainable_layers > 0:
            base_model.trainable = True
            # Freeze all layers except the top trainable_layers
            for layer in base_model.layers[:-trainable_layers]:
                layer.trainable = False
        
        # Add custom classification head optimized for MobileNet
        model = Sequential([
            base_model,
            GlobalAveragePooling2D(),
            Dense(256, activation='relu'),
            Dropout(0.3),
            Dense(128, activation='relu'),
            Dropout(0.2),
            Dense(num_classes, activation='softmax')
        ])
        
        return model
    
    def build_transfer_learning_model(self, num_classes, base_model_name='MobileNet'):
        """Build transfer learning model - now defaults to MobileNet only"""
        if base_model_name != 'MobileNet':
            print(f"Warning: Only MobileNet is supported. Using MobileNet instead of {base_model_name}")
        
        return self.build_mobilenet_model(num_classes)
    
    def train_model(self, model, train_gen, val_gen, epochs=50, model_name="groundnut_model"):
        """Train the model with callbacks"""
        print(f"\n=== TRAINING MODEL: {model_name} ===")
        
        # Compile model
        model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        print(model.summary())
        
        # Callbacks
        callbacks = [
            EarlyStopping(patience=10, restore_best_weights=True),
            ReduceLROnPlateau(factor=0.5, patience=5, min_lr=1e-7),
            ModelCheckpoint(f'{model_name}.h5', save_best_only=True)
        ]
        
        # Train model
        history = model.fit(
            train_gen,
            epochs=epochs,
            validation_data=val_gen,
            callbacks=callbacks,
            verbose=1
        )
        
        self.model = model
        self.history = history
        
        return history
    
    def evaluate_model(self, test_generator):
        """Evaluate model performance"""
        if self.model is None:
            print("No model found. Train a model first.")
            return
            
        print("\n=== MODEL EVALUATION ===")
        
        # Predictions
        test_generator.reset()
        predictions = self.model.predict(test_generator)
        y_pred = np.argmax(predictions, axis=1)
        y_true = test_generator.classes
        
        # Class names
        class_names = list(test_generator.class_indices.keys())
        
        # Accuracy
        accuracy = accuracy_score(y_true, y_pred)
        print(f"Test Accuracy: {accuracy:.4f}")
        
        # Classification report
        print("\nClassification Report:")
        print(classification_report(y_true, y_pred, target_names=class_names))
        
        # Confusion matrix
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=class_names, yticklabels=class_names)
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()
        
        return accuracy, y_pred, y_true
    
    def plot_training_history(self):
        """Plot training history"""
        if self.history is None:
            print("No training history found.")
            return
            
        fig, axes = plt.subplots(1, 2, figsize=(12, 4))
        
        # Accuracy
        axes[0].plot(self.history.history['accuracy'], label='Training Accuracy')
        axes[0].plot(self.history.history['val_accuracy'], label='Validation Accuracy')
        axes[0].set_title('Model Accuracy')
        axes[0].set_xlabel('Epoch')
        axes[0].set_ylabel('Accuracy')
        axes[0].legend()
        
        # Loss
        axes[1].plot(self.history.history['loss'], label='Training Loss')
        axes[1].plot(self.history.history['val_loss'], label='Validation Loss')
        axes[1].set_title('Model Loss')
        axes[1].set_xlabel('Epoch')
        axes[1].set_ylabel('Loss')
        axes[1].legend()
        
        plt.tight_layout()
        plt.show()
    
    def predict_single_image(self, image_path, class_names):
        """Predict a single image"""
        if self.model is None:
            print("No model found. Train a model first.")
            return
            
        img = load_img(image_path, target_size=self.img_size)
        img_array = img_to_array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)
        
        prediction = self.model.predict(img_array)
        predicted_class = class_names[np.argmax(prediction)]
        confidence = np.max(prediction)
        
        plt.figure(figsize=(8, 6))
        plt.imshow(img)
        plt.title(f'Predicted: {predicted_class} (Confidence: {confidence:.2f})')
        plt.axis('off')
        plt.show()
        
    def fine_tune_mobilenet(self, train_gen, val_gen, epochs=10):
        """Fine-tune MobileNet model by unfreezing some layers"""
        if self.model is None:
            print("No model found. Train a model first.")
            return
            
        print("\n=== FINE-TUNING MOBILENET ===")
        
        # Unfreeze some layers for fine-tuning
        base_model = self.model.layers[0]
        base_model.trainable = True
        
        # Fine-tune from this layer onwards
        fine_tune_at = len(base_model.layers) - 20
        
        # Freeze all the layers before fine_tune_at
        for layer in base_model.layers[:fine_tune_at]:
            layer.trainable = False
        
        # Use a lower learning rate for fine-tuning
        self.model.compile(
            optimizer=Adam(learning_rate=0.0001/10),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        print(f"Fine-tuning from layer {fine_tune_at} onwards...")
        print(f"Total trainable variables: {len(self.model.trainable_variables)}")
        
        # Callbacks for fine-tuning
        callbacks = [
            EarlyStopping(patience=5, restore_best_weights=True),
            ReduceLROnPlateau(factor=0.5, patience=3, min_lr=1e-8),
            ModelCheckpoint('mobilenet_finetuned.h5', save_best_only=True)
        ]
        
        # Fine-tune
        fine_tune_history = self.model.fit(
            train_gen,
            epochs=epochs,
            validation_data=val_gen,
            callbacks=callbacks,
            verbose=1
        )
        
        self.history = fine_tune_history
        
        return fine_tune_history

# Usage example and main analysis pipeline
def main():
    # Initialize analyzer
    analyzer = GroundnutLeafAnalyzer()
    
    # Explore dataset
    df_counts = analyzer.explore_dataset_structure()
    if df_counts is not None:
        analyzer.visualize_class_distribution(df_counts)
        analyzer.sample_images_visualization()
        analyzer.analyze_image_properties()
        
        # Create data generators
        train_gen, val_gen, test_gen = analyzer.create_data_generators()
        
        # Build and train MobileNet model only
        num_classes = train_gen.num_classes
        
        print("\n" + "="*50)
        print("TRAINING MOBILENET MODEL")
        print("="*50)
        
        # Build MobileNet model
        mobilenet_model = analyzer.build_mobilenet_model(num_classes)
        
        # Train the model
        history = analyzer.train_model(mobilenet_model, train_gen, val_gen, 
                                     epochs=20, model_name="mobilenet_groundnut")
        
        # Plot training history
        analyzer.plot_training_history()
        
        # Evaluate model
        accuracy, y_pred, y_true = analyzer.evaluate_model(test_gen)
        
        # Optional: Fine-tune the model for better performance
        print("\n" + "="*50)
        print("OPTIONAL: FINE-TUNING MOBILENET")
        print("="*50)
        
        fine_tune_response = input("Do you want to fine-tune the model? (y/n): ").lower()
        if fine_tune_response == 'y':
            fine_tune_history = analyzer.fine_tune_mobilenet(train_gen, val_gen, epochs=10)
            analyzer.plot_training_history()  # Plot fine-tuning history
            analyzer.evaluate_model(test_gen)  # Evaluate fine-tuned model
        
        print("\n=== MOBILENET ANALYSIS COMPLETE ===")
        print(f"Final Test Accuracy: {accuracy:.4f}")
        print("MobileNet model trained and evaluated successfully!")
        print("You can now extend this pipeline with additional datasets.")

if __name__ == "__main__":
    main()
```
